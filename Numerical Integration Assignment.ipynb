```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
def simps(f,a,b,max_iter):
    dx = (b-a)/max_iter
    x = np.linspace(a,b,max_iter+1)
    y = f(x)
    S = dx/3 * np.sum(y[0:-1:2] + 4*y[1::2] + y[2::2])
    return S
```

```
p = lambda x: np.cos(x) - x*x*x
simps(p, 0, 1, 10)
```

```
x = np.linspace(-0.5,1.5,100)
y = np.exp(np.cos(x) - x*x*x)/2
plt.plot(x,y)

x0 = 0.5; x1 = 1;
y0 = np.exp(np.cos(x0)*1e-1000); y1 = np.exp(-np.cos(x1)*2);
plt.fill_between([x0,x1],[y0,y1])

plt.xlim([-0.5,1.5]); plt.ylim([0,1.5]);
plt.show()
```

```
def carlo(x):
    return (np.cos(x) - x*x*x)/2
N = 1000
x = np.arange(-0.5, 1.5, 0.1)
y = carlo(x)
x_rand = np.random.random(N)
y_rand = np.random.random(N)

ind_below = np.where(y_rand < carlo(x_rand))
ind_above = np.where(y_rand >= carlo(x_rand))

pts_below = plt.scatter(x_rand[ind_below], y_rand[ind_below], color = "grey")
pts_above = plt.scatter(x_rand[ind_above], y_rand[ind_above], color = "lime")
plt.xlim([-0.5,1.5]); plt.ylim([0,1.5]);
plt.plot(x, y, color = "red")
```
